## In which year were the most runaways?

```python
df_freedom.info()
```

We can see that the "adv_publication_date" is not in data type date and it would be hard to extract informations, so in order to be able to analyze it we have to change the data type in date.

```python
df_freedom['adv_publication_date'] = pd.to_datetime(df_freedom['adv_publication_date'])
```

```python
df_freedom.info()
```

```python
plt.figure(figsize=(10, 6))
df_freedom['adv_publication_date'].dt.year.value_counts().sort_index().plot(kind='bar', rot = 90)
plt.title('Distribution of Publication Dates')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(fontsize=8)
plt.show()
```

```python
import seaborn as sns


plt.figure(figsize=(14, 8))
sns.lineplot(x='adv_publication_date', y='runaway_location_state', data=df_freedom, ci=None)
plt.title('Number of Runaways Over Time by State')
plt.xlabel('Publication Year')
plt.ylabel('Number of Runaways')
plt.show()
```
